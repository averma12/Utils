# Let's combine all the functions here to use as final implementation

```python
import os
import subprocess
import json
import time
import re
from IPython.display import Audio, display

def download_audio(url, output_dir="downloads"):
    # Step 2: Directory Creation
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    # Step 3: Command Construction
    command = [
        'yt-dlp',        # The program we're running
        '-x',            # Extract audio flag
        '--audio-format', 'mp3',  # Specify output format
        '-o', f'{output_dir}/%(title)s.%(ext)s',  # Where to save
        url              # The video URL
    ]
    result = subprocess.run(command, capture_output=True, text=True)
    return result
def extract_filepath(stdout_text):
    pattern = r'\[ExtractAudio\] Destination: (.+)'
    for line in stdout_text.splitlines():
        match = re.match(pattern, line.strip())
        if match:
            return match.group(1)
    return None
def verify_audio_file(file_path):
    """
    Basic verification: Check if file exists and has non-zero size
    Returns: Boolean indicating if verification passed
    """
    if os.path.exists(file_path):
        size_mb = os.path.getsize(file_path) / (1024*1024)
        print(f"File exists at: {file_path}")
        print(f"File size: {size_mb:.2f} MB")

        # Check if file has content (not empty)
        if size_mb > 0:
            return True
        else:
            print("File exists but is empty")
            return False

    print("File does not exist")
    return False


url = "https://www.youtube.com/live/HO1gmouGEVI"
output_dir = "."
```

```python
def download_and_verify_audio(url, output_dir="downloads"):
    result = download_audio(url, output_dir)
    if result.returncode == 0:
        file_path = extract_filepath(result.stdout)
        if verify_audio_file(file_path):
            return file_path
    return None
```

```python
def transcribe_audio(file_path):
    with open(file_path, "rb") as f:
        response = requests.post(
            "https://audio-prod.us-virginia-1.direct.fireworks.ai/v1/audio/transcriptions",
            headers={"Authorization": f"Bearer {FIREWORKS_API_KEY}"},
            files={"file": f},
            data={
                "vad_model": "silero",
                "alignment_model": "tdnn_ffn",
                "preprocessing": "none",
                "temperature": "0",
                "timestamp_granularities": "segment"
            },
        )

    if response.status_code == 200:
        print(response.json())
        return response.json()
    else:
        print(f"Error: {response.status_code}", response.text)
        return {"error": response.text}
```

```python
def seconds_to_timestamp(seconds):
    """Convert seconds to HH:MM:SS.mmm format"""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds_remainder = seconds % 60
    return f"{hours:02d}:{minutes:02d}:{seconds_remainder:06.3f}"
```

# Get transcription with segments

```python
def transcribe_audio_with_segments(file_path):
    with open(file_path, "rb") as f:
        response = requests.post(
            "https://audio-prod.us-virginia-1.direct.fireworks.ai/v1/audio/transcriptions",
            headers={"Authorization": f"Bearer {FIREWORKS_API_KEY}"},
            files={"file": f},
            data={
                "vad_model": "silero",
                "alignment_model": "tdnn_ffn",
                "preprocessing": "none",
                "temperature": "0",
                "timestamp_granularities": "segment",
                "response_format": "verbose_json"
            },
        )

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}", response.text)
        return {"error": response.text}
```

```python
def seconds_to_timestamp(seconds):
    """Convert seconds to HH:MM:SS.mmm format"""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds_remainder = seconds % 60
    return f"{hours:02d}:{minutes:02d}:{seconds_remainder:06.3f}"
```

```python
def create_timestamped_transcript(segments):
    """Convert segments to timestamped text with start and end times"""
    timestamped_lines = []

    for segment in segments:
        start_time = seconds_to_timestamp(segment['start'])
        end_time = seconds_to_timestamp(segment['end'])
        text = segment['text'].strip()
        timestamped_lines.append(f"[{start_time} -> {end_time}] {text}")

    return "\n".join(timestamped_lines)
```
